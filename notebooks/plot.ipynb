```python
import pandas as pd
import matplotlib.pyplot as plt
```

## All iterations


```python
df = pd.read_csv("../out/cost_history.csv")
plt.plot(df["iteration"], df["cost"])
plt.yscale("log")
plt.xlabel("Iteration")
plt.ylabel("Cost (log scale)")
plt.title("Cost Function Convergence")
plt.show()
```

## N first iterations
To change number of iterations change n variable

```python
n = 10
plt.plot(df["iteration"][:n], df["cost"][:n])
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.title("Cost Function Convergence (first 1000 iterations)")
plt.show()
```
